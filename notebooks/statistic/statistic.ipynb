```python
"""
    Count max quantity of relevant articles
"""
import json

with open("E:/Kaggle/icail/data/coliee-2021/coliee-2021-preprocessed/combined_en_jp_test_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

max_articles = 0
max_item_id = None

for item in data:
    num_articles = len(item.get("relevant_articles", []))

    if num_articles > max_articles:
        max_articles = num_articles
        max_item_id = item.get("id")

print("Số relevant_articles lớn nhất:", max_articles)
print("ID tương ứng:", max_item_id)
```
